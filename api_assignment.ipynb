## Q1. What is an API? Give an example where an API is used in real life.

**Answer:**
API stands for Application Programming Interface. It is a set of rules that allows different software applications to communicate with each other. An API works like a messenger that takes a request from one system and returns the response from another system.

**Example:**
When we use a weather app on our phone, the app sends a request to a weather server through an API. The server sends back weather data such as temperature, humidity, and forecast, which the app displays to the user.

## Q2. Give advantages and disadvantages of using API.

**Advantages:**
1. APIs allow different applications to communicate easily.
2. They help developers reuse existing services instead of building everything from scratch.
3. APIs speed up development time.
4. They allow integration between different platforms.

**Disadvantages:**
1. APIs may create security risks if not managed properly.
2. Dependency on external services can cause problems if the API stops working.
3. Some APIs have usage limits or require payment.

## Q3. What is a Web API? Differentiate between API and Web API.

**Web API:**
A Web API is an API that is accessed over the internet using HTTP or HTTPS protocols. It allows web applications and services to communicate with each other.

**Difference between API and Web API:**

| API | Web API |
|-----|--------|
| Can work locally within a system | Works over the internet |
| Not always web-based | Always web-based |
| May use different protocols | Usually uses HTTP/HTTPS |

## Q4. Explain REST and SOAP Architecture. Mention shortcomings of SOAP.

**REST (Representational State Transfer):**
REST is an architectural style used for designing web services. It uses standard HTTP methods such as GET, POST, PUT, and DELETE. REST APIs are simple, lightweight, and commonly use JSON format.

**SOAP (Simple Object Access Protocol):**
SOAP is a protocol used for exchanging structured information between systems. It uses XML format for sending messages.

**Shortcomings of SOAP:**
1. SOAP messages are large because they use XML.
2. It is slower compared to REST.
3. Implementation is more complex.
4. Requires more bandwidth.
